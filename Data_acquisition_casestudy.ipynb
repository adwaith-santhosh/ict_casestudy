Task 8

Merge all four datasets into one DataFrame.

Hint:

● Use the common column (country)

● Choose an appropriate join method.

```python
merged_data=pd.merge(data1,data2,on='country',how='inner')
merged_data=pd.merge(merged_data,data3,on='country',how='inner')
merged_data=pd.merge(merged_data,data4,on='country',how='inner')
merged_data
```

Task 9

Create the following calculated columns:

1. Internet Penetration (%)

2. GDP Per Capita

Hint:

Round the calculated values to two decimal places.

```python
merged_data['Internet penetration(%)']=((merged_data['internet_users']/merged_data['population'])*100).round(2)
merged_data['GDP Per Capita']=(merged_data['GDP']/merged_data['population']).round(2)
merged_data
```

Task 10

Create a SQLite database. Store the final merged dataset into a table named

‘country_statistics’

Hint:

● Use the sqlite3 library.

● Load the DataFrame using pandas.

```python
import sqlite3 as sq
```

```python
conn=sq.connect('country_statistics.db')
merged_data.to_sql('country_statistics',conn,if_exists='replace',index=False)
print('database created successfully')
```

Part F – Data Analysis Using SQL

Answer the following questions using SQLite queries.

11. Display the Top 5 countries by GDP.

```python
q11='''SELECT country, GDP FROM country_statistics ORDER BY GDP DESC LIMIT 5;'''
pd.read_sql(q11,conn)

```

12. Find the average literacy rate.

```python
q12='''SELECT AVG(literacy_rate) from country_statistics;  '''
pd.read_sql(q12,conn)
```

13. Display all countries having literacy rate above 90%.

```python
q13 = """SELECT country, literacy_rate FROM country_statistics WHERE literacy_rate > 90;"""

print(pd.read_sql(q13, conn))
```

14. Find the Top 5 countries with the highest internet penetration.

```python
q14 = """
SELECT country, [Internet penetration(%)]
FROM country_statistics
ORDER BY [Internet penetration(%)] DESC
LIMIT 5;
"""

print(pd.read_sql(q14, conn))
```

15. Calculate the total population

```python
q15 = """
SELECT SUM(population) AS total_population
FROM country_statistics;
"""

print(pd.read_sql(q15, conn))
```

16. Find the country with the highest GDP per capita.  

```python
q16 = """
SELECT country, [GDP Per Capita]
FROM country_statistics
ORDER BY [GDP Per Capita] DESC
LIMIT 1;
"""

print(pd.read_sql(q16, conn))
```
